```python
print("Categories in Gender feature:",end="")
print(len(df['gender'].unique()))
print("Categories in race_ethnicity feature:",end="")
print(len(df['race_ethnicity'].unique()))
print('Categories in parental_level_of_education feature:',end=" ")
print(len(df['parental_level_of_education'].unique()))
print("Categories in lunch feature:",end="")
print(len(df['lunch'].unique()))
print("Categories in test_preparation_course feature:",end="")
print(len(df['test_preparation_course'].unique()))
```

```python
numerical_features = df.select_dtypes(exclude='object').columns
categorical_features = df.select_dtypes(include='object').columns
print("Numerical Features:",end=" ")
print(len(numerical_features))
print("Categorical Features:",end=" ")
print(len(categorical_features))
print("Total Features:",end=" ")
print(len(numerical_features)+len(categorical_features))
print("Total Features in dataset:",end=" ")
print(len(df.columns))

```

```python
df['total_score'] = df['math_score'] + df['reading_score'] + df['writing_score']
df['average'] = df['total_score'] / 3
```

```python
df.head()
```

```python
reading_full = df[df['reading_score'] == 100]
reading_full
```

```python
reading_full = df[df['reading_score'] == 100]['total_score'].count()
reading_full
```

```python
df.shape
```

```python
reading_full = df[df['reading_score'] == 100]['average'].count()
writing_full = df[df['writing_score'] == 100]['average'].count()
math_full = df[df['math_score'] == 100]['average'].count()

print(f'Number of students with full marks in Maths: {math_full}')
print(f'Number of students with full marks in Writing: {writing_full}')
print(f'Number of students with full marks in Reading: {reading_full}')
```

```python
reading_less_20 = df[df['reading_score'] <= 20]['average'].count()
writing_less_20 = df[df['writing_score'] <= 20]['average'].count()
math_less_20 = df[df['math_score'] <= 20]['average'].count()

print(f'Number of students with less than 20 marks in Maths: {math_less_20}')
print(f'Number of students with less than 20 marks in Writing: {writing_less_20}')
print(f'Number of students with less than 20 marks in Reading: {reading_less_20}')
```

```python

plt.figure(figsize=(18,8))
plt.subplot(1, 4, 1)
plt.title('MATH SCORES')
sns.violinplot(y='math_score',data=df,color='red',linewidth=3,inner='point')
plt.subplot(1, 4, 2)
plt.title('READING SCORES')
sns.violinplot(y='reading_score',data=df,color='green',linewidth=3,inner='point')
plt.subplot(1, 4, 3)
plt.title('WRITING SCORES')
sns.violinplot(y='writing_score',data=df,color='blue',linewidth=3,inner='point')
plt.show()
```

```python
plt.rcParams['figure.figsize']=(20,10)
plt.subplot(1,5,1)
value=df['gender'].value_counts()
colors=['red','green']
labels='female','male'

plt.title("male vs female")
plt.pie(value,labels=labels,colors=colors,autopct='%.2f%%')
plt.axis('off')
plt.show()
```

```python
f,ax=plt.subplots(1,2,figsize=(20,10))
sns.countplot(x=df['gender'],data=df,palette ='bright',ax=ax[0],saturation=0.95)
for container in ax[0].containers:
    ax[0].bar_label(container,color='black',size=20)
    
plt.pie(x=df['gender'].value_counts(),labels=['Male','Female'],explode=[0,0.1],autopct='%1.1f%%',shadow=True,colors=['#ff4d4d','#ff8000'])
plt.show()
```

```python
plt.figure(figsize=(10, 8))

X = ['Total Average','Math Average','reading_score']


female_scores = [gender_group['average'][0], gender_group['math_score'][0],gender_group['reading_score'][0]]
male_scores = [gender_group['average'][1], gender_group['math_score'][1],gender_group['reading_score'][1]]


X_axis = np.arange(len(X))
  

plt.bar(X_axis+0.2, female_scores, 0.4, label = 'Female')
plt.bar(X_axis-0.2, male_scores, 0.4, label = 'Male')
  
plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total average v/s Math average marks of both the genders", fontweight='bold')
plt.legend()
plt.show()
```
